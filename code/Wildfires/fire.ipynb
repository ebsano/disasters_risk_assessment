```python
import pandas as pd
```

```python
active_fires_url = 'https://www.fire.ca.gov/umbraco/api/IncidentApi/List?inactive=flase'
active_fires = pd.read_json(active_fires_url)
inactive_fires_url = 'https://www.fire.ca.gov/umbraco/api/IncidentApi/List?inactive=true'
inactive_fires = pd.read_json(inactive_fires_url)
fires_2020 = inactive_fires[inactive_fires['ExtinguishedDateOnly'].str.contains('2020')]
fires = pd.concat([active_fires, fires_2020])
```

```python
fires.info()
```

```python
#finding null values
fires.isnull().sum()
```

```python
#dropping columns with so many null values
fires.drop(columns=['AdminUnitUrl', 'ControlStatement'], inplace=True)
```
