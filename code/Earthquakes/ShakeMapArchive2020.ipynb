```python
table2 = pd.read_html('https://www.cisn.org/shakemap/nc/shake/archive/',)
```

```python
pretty2 = table2[-1]
pretty2
```

```python
pretty2 = pretty2.iloc[1:].rename(columns = pretty2.iloc[0,:])
pretty2
```

```python
table3 = pd.read_html('https://www.cisn.org/shakemap/sc/shake/archive/',)
```

```python
pretty3 = table3[-1]
pretty3
```

```python
pretty3 = pretty3.iloc[1:].rename(columns = pretty3.iloc[0,:])
pretty3
```

```python
all_pretty = pretty.append(pretty2.append(pretty3)).reset_index(drop=True)
```

```python
all_pretty.sort_values(by='Event ID')
```

```python
all_pretty.duplicated(subset='Event ID')
```

```python
all_pretty.loc[all_pretty.duplicated(subset='Event ID')]
```

```python
single = all_pretty.drop_duplicates(subset='Event ID').copy()
```

```python
single.info()
```

```python
#NEXT STEP -- convert dtypes
```

```python
single['Date'] = pd.to_datetime(single['Date'])
```

```python
'%I:%M:%S %p %Z' 
```

```python
datetime.strptime(single.loc[single['Time'].str.contains('PST')], '%I:%M:%S %p %Z')
```

```python
for i in single['Time']:
    pd.to_datetime(i, tzinf)
```

```python
single.info()
```

```python
single['Lat'] = pd.to_numeric(single['Lat'])
```

```python
single['Lon'] = pd.to_numeric(single['Lon'])
```

```python
single['Mag'] = pd.to_numeric(single['Mag'])
```

```python
single.info()
```

```python
single
```

```python
single.drop('Time', axis =1 , inplace=True)
```

```python
single.head()
```

```python
single.to_csv('../../data/earthquakes_CA.csv', index=False)
```
